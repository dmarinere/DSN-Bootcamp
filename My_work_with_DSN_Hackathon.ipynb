```
dropcols = ['Applicant_ID', 'default_status']
y = train['default_status'] # labels
x = train.drop(columns=dropcols, axis=1)
test = test.drop(columns=['Applicant_ID'], axis=1) #will use this for predicting and submitting the resulting
print(x.shape)
print(y.shape)
print(test.shape)
```

```
from catboost import CatBoostClassifier
```

```
cate_features_index=np.where(x.dtypes != float) [0]; cate_features_index
```

```
from sklearn.metrics import accuracy_score,confusion_matrix,roc_auc_score, log_loss
errcb=[]
y_pred_totcb=[]
from sklearn.model_selection import KFold,StratifiedKFold, TimeSeriesSplit, GroupKFold
fold=StratifiedKFold(n_splits=5,shuffle=True, random_state=12345)
i=1
for train_index, test_index in fold.split(x,y):
    X_train, X_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    m=CatBoostClassifier(n_estimators=10000, max_depth=6,  bootstrap_type='Bernoulli',
                              subsample=0.9,
                              eval_metric='AUC',
                              metric_period=20,)
    m.fit(X_train,y_train,cat_features=cate_features_index,eval_set=[(X_train, y_train),(X_test, y_test)],verbose=100, early_stopping_rounds=200)#,verbose=100)
    preds=m.predict_proba(X_test)[:, 1]
    print("err: ",log_loss(y_test,preds))
    print("err: ",roc_auc_score(y_test,preds))
    errcb.append(log_loss(y_test,preds))
    p = m.predict_proba(test)[:, 1]
    y_pred_totcb.append(p)
```

```
d = {"Applicant_ID": test_id, 'default_status': np.mean(y_pred_totcb, 0)}
test_predictions = pd.DataFrame(data=d)
test_predictions = test_predictions[["Applicant_ID", 'default_status']]
```

```
test_predictions.to_csv('25kfoldwith2train.csv', index=False)
```
